```python
from PIL import Image
import os
import numpy as np
import pandas as pd
```

```python
img = np.array(Image.open('Image/m_3009743_sw_14_060_20201018.tif'))
img.shape
```

```python
def find_area(img, x, y, d=1):
    return img[max(0, x-d):x+d+1, max(0, y-d):y+d+1].astype(int)
```

```python
x = pd.DataFrame({'x':np.random.choice(img.shape[0], 500)})
y = pd.DataFrame({'y':np.random.choice(img.shape[1], 500)})

# cross join
df = x.assign(foo = 1).merge(y.assign(foo = 1), on = 'foo').drop(columns = ['foo'])
```

```python
df[['r', 'g', 'b', 'l']] = pd.DataFrame(df.apply(
    lambda x:img[int(x['x']), int(x['y'])].astype(int), axis = 1).tolist())

df[['r_avg', 'g_avg', 'b_avg', 'l_avg']] = pd.DataFrame(
    df.apply(lambda x:np.average(find_area(img, int(x['x']), int(x['y']), 2), axis = (0,1)), axis = 1).tolist())

df[['r_std', 'g_std', 'b_std', 'l_std']] = pd.DataFrame(
    df.apply(lambda x:np.std(find_area(img, int(x['x']), int(x['y']), 2), axis = (0,1)), axis = 1).tolist())
```
